```python
#Problem 8
df7=Data[['dt','member_id']]
df7_1=df7[df7.dt=='2019-01-20'].drop_duplicates(['member_id']) #the table of 2019-01-20 of unique users
df7_2=df7[df7.dt=='2019-01-21'].drop_duplicates(['member_id'])
df7_3=df7[df7.dt=='2019-01-22'].drop_duplicates(['member_id'])
df7_4=df7[df7.dt=='2019-01-23'].drop_duplicates(['member_id'])
df7_5=df7[df7.dt=='2019-01-24'].drop_duplicates(['member_id'])
df7_6=df7[df7.dt=='2019-01-25'].drop_duplicates(['member_id'])
df7_7=df7[df7.dt=='2019-01-26'].drop_duplicates(['member_id'])
df7_8=df7[df7.dt=='2019-01-27'].drop_duplicates(['member_id'])
df7_1_next=pd.concat([df7_2,df7_3,df7_4,df7_5,df7_6,df7_7,df7_8])
Retention_Rate_Week=len(df7_1['member_id'].loc[df7_1['member_id'].isin(df7_1_next['member_id'])])/len(df7_1['member_id'])
Retention_Rate_Week

#another way
#define the function calculating the Retention_Rate_Week
from datetime import datetime
from datetime import timedelta
def Retention_Rate_Week(x,y=7):  #x:the beginning date,y:the interval;default=7
    df7=Data[['dt','member_id']]
    df7_x=df7[df7.dt==x].drop_duplicates(['member_id'])
    
    i=1
    while i<=y:
        x=datetime.strptime(x,'%Y-%m-%d')+timedelta(1) #str to timeseries and move one day
        x=x.strftime('%Y-%m-%d') #timeseries to str
        df7_y=df7[df7.dt==x].drop_duplicates(['member_id'])
        if i==1:
            df7_z=df7_y
        else:
            df7_z=pd.concat([df7_y,df7_z])
        i=i+1
        
    Retention_Rate_xy=len(df7_x['member_id'].loc[df7_x['member_id'].isin(df7_z['member_id'])])/len(df7_x['member_id'])
    return Retention_Rate_xy
Retention_Rate_Week('2019-01-20')
```

```python
#Problem 13
ActiveUser.describe() #give a brief introduction

import math
data=list(ActiveUser['total orders'])

def calc(data):   
    n=len(data)
    niu=0.0 #mean
    niu2=0.0 #quadratic mean
    niu3=0.0 #cubic mean
    for a in data:
        niu+=a
        niu2+=a**2
        niu3+=a**3
    niu/=n
    niu2/=n
    niu3/=n
    std=math.sqrt(niu2-niu*niu)
    var=std**2
    return [niu,std,var,niu3]

def calc_stat(data):   #calculate std,var,skew,kurt
    [niu,std,var,niu3]=calc(data)
    n=len(data)
    niu4=0.0
    for a in data:
        a-=niu
        niu4+=a**4
    niu4/=n
    
    skew=(niu3-3*niu*var-niu**3)/(std**3)
    kurt=niu4/(std**4)
    return [std,var,skew,kurt]

calc_stat(data)
```

```python
#Problem 3.a
#Xgboost
import xgboost as xgb
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split

#get training and test sets
X_train_input,X_test_input,Y_train_input,Y_test_true=train_test_split(X,Y,test_size=0.3,random_state=42)

dtrain=xgb.DMatrix(X_train_input,Y_train_input)
dtest=xgb.DMatrix(X_test_input)

#set the model
params={'booster':'gbtree',
    'objective': 'binary:logistic',
    'eval_metric': 'auc',
    'max_depth':4,
    'lambda':10,
    'subsample':0.75,
    'colsample_bytree':0.75,
    'min_child_weight':2,
    'eta': 0.025,
    'seed':0,
    'nthread':8,
     'silent':1}
watchlist = [(dtrain,'train')]

#build the model and predict
bst=xgb.train(params,dtrain,num_boost_round=100,evals=watchlist)

ypred=bst.predict(dtest)

#set the threshold
y_pred = (ypred >= 0.5)*1

from sklearn import metrics
print('AUC: %.4f' % metrics.roc_auc_score(Y_test_true,ypred))
print('ACC: %.4f' % metrics.accuracy_score(Y_test_true,y_pred))
print('Recall: %.4f' % metrics.recall_score(Y_test_true,y_pred))
print('F1-score: %.4f' %metrics.f1_score(Y_test_true,y_pred))
print('Precesion: %.4f' %metrics.precision_score(Y_test_true,y_pred))
metrics.confusion_matrix(Y_test_true,y_pred)
```

```python
#from the bar graph，the data is imbalanced,therefore we can do some oversampling

from imblearn.over_sampling import RandomOverSampler,SMOTE
from imblearn.combine import SMOTETomek

ros=RandomOverSampler(random_state=0) #随机过采样

X_ros,Y_ros=ros.fit_sample(X_train_input,Y_train_input)

print('ros:%s'%(len(Y_ros)))
```

```python
#Adaboost
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV


import warnings
warnings.filterwarnings("ignore")

data=[[X_ros,Y_ros],
     [X_train_input,Y_train_input]]

AdaBoost_clf = AdaBoostClassifier(base_estimator=None,algorithm='SAMME.R', random_state=0)  
param_grid={'n_estimators':[10,50],'learning_rate':[0.7,1]}
cv1=GridSearchCV(AdaBoost_clf,param_grid=param_grid,scoring='f1')

for features,labels in data:
    cv1.fit(features,labels)
    Y_prediction = cv1.predict(X_test_input)
    print(classification_report(Y_test_true,Y_prediction))

print(cv1.best_params_)
```
